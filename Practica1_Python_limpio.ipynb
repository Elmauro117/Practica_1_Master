```python
def Merge(sorted_left_array, sorted_right_array):
    
    # El nuevo array en donde vamos a appendear.
    arr_resultado = []
    
    # Itermaos los dos arrays 
    while len(sorted_left_array) > 0 and len(sorted_right_array) > 0:
        if sorted_left_array[0] > sorted_right_array[0]:
            arr_resultado.append(sorted_right_array[0])
            sorted_right_array.pop(0)
        else:
            arr_resultado.append(sorted_left_array[0])
            sorted_left_array.pop(0)
    
    while len(sorted_left_array) > 0:
        arr_resultado.append(sorted_left_array[0])
        sorted_left_array.pop(0)

    while len(sorted_right_array) > 0:
        arr_resultado.append(sorted_right_array[0])
        sorted_right_array.pop(0)
    
    return arr_resultado

def mergeSort(arr):
    #Primero miramos si se puede seguir dividiendo el problema
    if len(arr) == 1:
        return arr

    #Si todavia se puede dividir, se divide en dos
    middle = len(arr) // 2
    left_array = arr[:middle]
    right_array = arr[middle:]

    # Aplicas MergeSort recursivamente
    sorted_left_array = mergeSort(left_array)
    sorted_right_array = mergeSort(right_array)

    # Finalizamos aplicando la func MERGE a los dos array declarados arriba
    return Merge(sorted_left_array, sorted_right_array)

```

```python
# Algoritmo Bubble Sort

def burble_sort(Lista):
    
    # Algoritmo
    swapped = False
    for i in range(len(Lista)-1):
        for j in range(0,len(Lista)-i-1):
            if Lista[j]> Lista[j+1]:
                swapped = True
                Lista[j],Lista[j+1]=Lista[j+1],Lista[j]
                      
        if not swapped:
            
            return 0
    return Lista
```

```python

# Algoritmo Selection Sort
def selection_sort(Lista):
   
    # Algoritmo de ordenacion
    for i in range(len(Lista)-1):
        for j in range(i+1,len(Lista)):
            if Lista[i]>Lista[j]:
                Lista[i],Lista[j]=Lista[j],Lista[i]
                
    return Lista
                
                
# Algoritmo Counting Sort
def counting_sort(Lista):
          
    o = [0 for i in range(len(Lista))]   # tiene el tamaÃ±o del arreglo original  LISTA
    count = [0 for i in range(max(Lista)+1)] # arreglo con los numeros que llegan hasta el numero mas grande de la lista

    #contamos cuantas veces aparece cada elemento del arreglo
    for i in Lista:
        count[i] += 1
      

    #suma acumulativa, en el rango debe ir la cantidad de numeros q hay en el arreglo de la lista (hasta el num mas grand de la lista)
    for i in range(1,max(Lista)+1):
        count[i] += count[i-1]
       

    #ordenamos segun los indices
    for i in range(len(Lista)):
        o[count[Lista[i]]-1]=Lista[i]
        count[Lista[i]]-=1
    return o
```
